```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
Hesi_data = "C:\\Users\\0\\Desktop\\출산꺼리는이유.csv"
Hesi = pd.read_csv(Hesi_data, encoding='cp949', index_col='전체')

# 전체 응답자의 출산관련 이유 중 가장 많이 선택된 이유부터 정렬
sorted_Hesi = Hesi.iloc[:, 1:].loc['평균'].sort_values(ascending=False)

plt.rcParams['font.size'] = 12

# Pie chart 그리기
plt.figure(figsize=(12, 8))

# 비율 계산
labels = sorted_Hesi.index
sizes = sorted_Hesi.values

# 가장 큰 값 3개의 인덱스 가져오기
explode_idx = sorted_Hesi.head(2).index

# explode 설정
explode = [0.1 if label in explode_idx else 0 for label in labels]

# 색상 설정
colors = sns.color_palette("Set3", len(labels))

# 차트 그리기
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow = True, wedgeprops=dict(width=0.7, edgecolor='w'))
plt.title('출산을 꺼리는 이유',fontsize=20, fontweight='bold')


plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 출산율 데이터 불러오기
Birth_data = "C:\\Users\\0\\Desktop\\출산율.csv"
Birth_data = pd.read_csv(Birth_data, encoding='cp949')

# 데이터 전처리
Birth = Birth_data.drop(index=0)
Birth.set_index('시군구별', inplace=True)

for j in Birth.columns:
    try:
        Birth[j] = Birth[j].astype(float)
    except ValueError as e:
        print(f"Error converting column {j} to float")

# 그래프 그리기
plt.figure(figsize=(16, 6))

# 막대그래프 그리기
ax = sns.barplot(x=Birth.columns, y=Birth.loc['세종특별자치시'], color='#4682B4', alpha=0.7, label='세종특별자치시')  # SteelBlue 색상
ax = sns.barplot(x=Birth.columns, y=Birth.loc['전국'], color='#4169E1', alpha=0.4, label='전국')  # RoyalBlue 색상
ax = sns.barplot(x=Birth.columns, y=Birth.loc['서울특별시'], color='#FF8C00', label='서울특별시')  # DarkOrange 색상

plt.ylim(0, 2)

plt.xticks(fontsize=10, rotation=45)
plt.xlabel('연도별', fontsize=15)
plt.ylabel('출산율(명)', fontsize=15)
plt.title('서울특별시 & 세종특별자치시 출산율 비교', fontsize=20)

# 범례 추가 및 색상 설정
legend = plt.legend()
legend.legendHandles[0].set_color('SteelBlue')
legend.legendHandles[1].set_color('Navy') 
legend.legendHandles[2].set_color('#FF8C00') 



plt.show()
```

```python
# 1. 출산율 & 총소득
Data = pd.read_csv("C:\\Users\\0\\Desktop\\출산율_총소득.csv", encoding='cp949')

# 모든 열을 float로 변환
for j in Data.columns:
    try:
        Data[j] = Data[j].astype(float)
    except ValueError as e:
        print()

# '연도별' 열을 인덱스로 설정
Data.set_index('연도별', inplace=True)

# 첫 번째 y 축에 대한 그래프 (막대 그래프)
fig, ax1 = plt.subplots(figsize=(16, 6))

# 첫 번째 y 축에 대한 막대 그래프
ax = sns.barplot(x=Data.columns, y=Data.loc['전국출산율'], ax=ax1, color='#3498db')

ax.set_ylabel('전국출산율', color='black', fontsize=16)
#plt.ylim(150, 350)

# x 축 눈금의 폰트 크기와 회전 설정
plt.xticks(fontsize=10)

# 두 번째 y 축 생성
ax2 = ax1.twinx()

# 두 번째 y 축에 대한 선 그래프
ax2.plot(Data.columns, Data.loc['전국총소득'], color='#e74c3c', marker='o')

# 두 번째 y 축의 눈금 설정
#ax2.set_ylim(40, 50)  # 두 번째 y 축의 범위를 0부터 100으로 설정
ax2.set_ylabel('1인당 평균소득(천원)', color='black', fontsize=12)

ax.xaxis.grid(True)
ax.yaxis.grid(True)

plt.title('출산울 & 1인당 평균소득',fontsize=20, pad=10)
```
