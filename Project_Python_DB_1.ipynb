```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
Hesi_data = "C:\\Users\\0\\Desktop\\.csv"
Hesi = pd.read_csv(Hesi_data, encoding='cp949', index_col='')

#          
sorted_Hesi = Hesi.iloc[:, 1:].loc[''].sort_values(ascending=False)

plt.rcParams['font.size'] = 12

# Pie chart 
plt.figure(figsize=(12, 8))

#  
labels = sorted_Hesi.index
sizes = sorted_Hesi.values

#    3  
explode_idx = sorted_Hesi.head(2).index

# explode 
explode = [0.1 if label in explode_idx else 0 for label in labels]

#  
colors = sns.color_palette("Set3", len(labels))

#  
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow = True, wedgeprops=dict(width=0.7, edgecolor='w'))
plt.title('  ',fontsize=20, fontweight='bold')


plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#   
Birth_data = "C:\\Users\\0\\Desktop\\.csv"
Birth_data = pd.read_csv(Birth_data, encoding='cp949')

#  
Birth = Birth_data.drop(index=0)
Birth.set_index('', inplace=True)

for j in Birth.columns:
    try:
        Birth[j] = Birth[j].astype(float)
    except ValueError as e:
        print(f"Error converting column {j} to float")

#  
plt.figure(figsize=(16, 6))

#  
ax = sns.barplot(x=Birth.columns, y=Birth.loc[''], color='#4682B4', alpha=0.7, label='')  # SteelBlue 
ax = sns.barplot(x=Birth.columns, y=Birth.loc[''], color='#4169E1', alpha=0.4, label='')  # RoyalBlue 
ax = sns.barplot(x=Birth.columns, y=Birth.loc[''], color='#FF8C00', label='')  # DarkOrange 

plt.ylim(0, 2)

plt.xticks(fontsize=10, rotation=45)
plt.xlabel('', fontsize=15)
plt.ylabel('()', fontsize=15)
plt.title(' &   ', fontsize=20)

#     
legend = plt.legend()
legend.legendHandles[0].set_color('SteelBlue')
legend.legendHandles[1].set_color('Navy') 
legend.legendHandles[2].set_color('#FF8C00') 



plt.show()
```
