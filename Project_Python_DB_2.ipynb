```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 생성
data = {
    '항목': ['반드시 해야 한다', '하는 편이 좋다', '해도 좋고 하지 않아도 좋다', '하지 않는 게 낫다', '모르겠다'],
    '2021': [8.576, 37.684, 46.328, 5.428, 1.992]
}

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 데이터를 '2021' 컬럼 기준으로 내림차순 정렬
df = df.sort_values('2021', ascending=False)

# 가장 큰 값 2개를 강조하기 위한 explode 설정
explode = [0.1 if i < 2 else 0 for i in range(df.shape[0])]

# seaborn의 "Pastel1" 색상 팔레트 적용
colors = sns.color_palette("Pastel1", len(df))

# 피규어 크기 설정
plt.figure(figsize=(10, 10))

# 파이차트 생성
plt.pie(df['2021'], labels=df['항목'], autopct='%1.1f%%', startangle=90, explode=explode, shadow=True, colors=colors, wedgeprops=dict(width=0.7, edgecolor='w'), textprops={'fontsize': 13})
plt.title('출산에 대한 의사', fontsize=20, fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 남성과 여성의 연령별 인구 데이터 예시
ages = np.arange(0, 101, 10)
male_population = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600]
female_population = [90, 120, 180, 220, 280, 320, 370, 410, 460, 510, 550]

# 맷플롯립을 사용하여 인구 피라미드 그리기
fig, ax = plt.subplots()

# 남성 인구 피라미드
ax.barh(ages, male_population, color='blue', label='Male')

# 여성 인구 피라미드 (음수값으로 설정하여 반대 방향으로 그림)
ax.barh(ages, [-x for x in female_population], color='pink', label='Female')

ax.set_xlabel('Population')
ax.set_ylabel('Age')
ax.set_title('Population Pyramid')
ax.legend()

# y 축 역순으로 설정하여 위에서 아래로 연령이 증가하도록 함
ax.invert_yaxis()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 생성
data = {
    '연도별': [2017, 2018, 2019, 2020, 2021],
    '예산': [24.1, 26.3, 32.4, 40.2, 46.7],
    '출산율': [1.05, 0.98, 0.92, 0.84, 0.81]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 피규어 크기 설정
fig, ax1 = plt.subplots(figsize=(10, 6))

# 출산율 막대그래프
ax1.bar(df['연도별'], df['출산율'], color='skyblue', label='출산율')

# 왼쪽 y축 레이블
ax1.set_ylabel('출산율', color='skyblue')

# 오른쪽 y축을 위한 축 생성
ax2 = ax1.twinx()

# 예산 선그래프
ax2.plot(df['연도별'], df['예산'], marker='o', color='orange', linestyle='-', linewidth=2, label='예산')

# 오른쪽 y축 레이블
ax2.set_ylabel('예산', color='orange')

# 타이틀 설정
plt.title('연도별 출산율과 예산',fontsize=20)

# 범례 추가
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='lower left', bbox_to_anchor=(0.8, 1))

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 주어진 데이터
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]).reshape(-1, 1)
birth_rate = np.array([1.205, 1.239, 1.172, 1.052, 0.977, 0.918, 0.837, 0.808, 0.778])
dual_income_ratio = np.array([44.2, 44.2, 45.3, 44.3, 46, 45.5, 45, 45.9, 46.1])

# 출산율과 혼인대비 맞벌이 비율 간의 관계 시각화
plt.figure(figsize=(10, 6))
plt.scatter(birth_rate, dual_income_ratio, color='blue', marker='o', label='실제 데이터')

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(birth_rate.reshape(-1, 1), dual_income_ratio)

# 회귀선을 위한 예측
birth_rate_pred = np.linspace(min(birth_rate), max(birth_rate), 100).reshape(-1, 1)
dual_income_ratio_pred = model.predict(birth_rate_pred)

# 회귀선 그리기
plt.plot(birth_rate_pred, dual_income_ratio_pred, color='red', linestyle='--', label='회귀선')

# 그래프에 제목과 축 레이블 추가
plt.title('출산율 & 맞벌이 비율 산점도와 회귀선', fontsize=16)
plt.xlabel('출산율(%)', fontsize=14)
plt.ylabel('맞벌이 비율(%)', fontsize=14)

# 각 데이터 포인트에 해당하는 연도 추가
for i, txt in enumerate(years):
    plt.annotate(txt[0], (birth_rate[i], dual_income_ratio[i]), textcoords="offset points", xytext=(0,10), ha='center')

# 그래프에 범례 추가
plt.legend()

# 그래프 표시
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 주어진 데이터
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]).reshape(-1, 1)
real_estate_index = np.array([3.84521675, 3.037258167, 0.787208917, -0.20503075, -0.92514225, 1.9760515, 9.92998825, 5.209869551, -10.99688478])
birth_rate = np.array([1.205, 1.239, 1.172, 1.052, 0.977, 0.918, 0.837, 0.808, 0.778])

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(birth_rate.reshape(-1, 1), real_estate_index)

# 회귀선을 위한 예측
birth_rate_pred = np.linspace(min(birth_rate), max(birth_rate), 100).reshape(-1, 1)
real_estate_index_pred = model.predict(birth_rate_pred)

# 산점도와 회귀선 그리기
plt.figure(figsize=(10, 6))
plt.scatter(birth_rate, real_estate_index, color='blue', marker='o', label='실제 데이터')
plt.plot(birth_rate_pred, real_estate_index_pred, color='red', linestyle='--', label='회귀선')

# 그래프에 제목과 축 레이블 추가
plt.title('출산율 & 매매지수 산점도와 회귀선', fontsize=16)
plt.xlabel('출산율(%)', fontsize=14)
plt.ylabel('매매지수', fontsize=14)

# 각 데이터 포인트에 해당하는 연도 추가
for i, txt in enumerate(years):
    plt.annotate(txt[0], (birth_rate[i], real_estate_index[i]), textcoords="offset points", xytext=(0,10), ha='center')

# 그래프에 범례 추가
plt.legend()

# 그래프 표시
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 주어진 데이터
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]).reshape(-1, 1)
birth_rate = np.array([1.205, 1.239, 1.172, 1.052, 0.977, 0.918, 0.837, 0.808, 0.778])
income_per_capita = np.array([17006, 18063, 18520, 19204, 19910, 20661, 21342, 22059, 23388])

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(birth_rate.reshape(-1, 1), income_per_capita)

# 회귀선을 위한 예측
birth_rate_pred = np.linspace(min(birth_rate), max(birth_rate), 100).reshape(-1, 1)
income_per_capita_pred = model.predict(birth_rate_pred)

# 산점도와 회귀선 그리기
plt.figure(figsize=(10, 6))
plt.scatter(birth_rate, income_per_capita, color='blue', marker='o', label='실제 데이터')
plt.plot(birth_rate_pred, income_per_capita_pred, color='red', linestyle='--', label='회귀선')

# 그래프에 제목과 축 레이블 추가
plt.title('출산율 & 1인당 개인소득 산점도와 회귀선', fontsize=16)
plt.xlabel('출산율(%)', fontsize=14)
plt.ylabel('1인당 개인소득(천원)', fontsize=14)

# 각 데이터 포인트에 해당하는 연도 추가
for i, txt in enumerate(years):
    plt.annotate(txt[0], (birth_rate[i], income_per_capita[i]), textcoords="offset points", xytext=(0,10), ha='center')

# 그래프에 범례 추가
plt.legend()

# 그래프 표시
plt.grid(True)
plt.show()
```
